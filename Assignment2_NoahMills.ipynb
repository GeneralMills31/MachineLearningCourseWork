Elbow method with IRIS dataset

```python
# We must utilize the elbow method to determine the number of clusters we have (Or the best guess at it)
# and then compare this value to the number of classes/clusters that are actually in the MNIST and IRIS datasets.
# IRIS has 3 Classes

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

iris = load_iris()

# Inertia is the sum of squared distances of each data point to its closest
# cluster center. Lower generally means better clustering.
# This code will loop through the values 1-10 and, for each, we will initalize
# a k-means algorithm with them as the k value.
# We will then use that current algorithm and fit it using the IRIS dataset.
# The inertia will calculated each time which will then be graphed and
# hopefully we will be able to see which k value is best.
inertia = []
k_values = range(1, 11)
for k in k_values:
  kmeans = KMeans(n_clusters=k, random_state=0)
  kmeans.fit(iris.data)
  inertia.append(kmeans.inertia_)

plt.plot(k_values, inertia, 'bo-')
plt.xlabel('Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method (IRIS Dataset)')
plt.show()
print("ACTUAL Number of classes in IRIS dataset: 3")
```
