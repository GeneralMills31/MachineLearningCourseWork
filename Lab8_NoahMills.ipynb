```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
plt.style.use('ggplot')

# Load data
data = load_breast_cancer()

# Explain classes, features, and samples/data.
print("CLASSES: \n\n", data.target_names)
print("\nFEATURES: \n\n", data.feature_names)
print("\nDESCRIPTION: \n\n", data.DESCR)

x = data.data
y = data.target
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

# GNB Testing
gnb = GaussianNB()
gnb.fit(x_train, y_train)
y_pred_gnb = gnb.predict(x_test)
print("\nGaussianNB Accuracy:", accuracy_score(y_test, y_pred_gnb))

# KNearest Neighbor Testing
knn = KNeighborsClassifier(9) # k = 9 Returns the best value (3 = .941, 5 = .959, 7 = .964, 9 = .97)
knn.fit(x_train, y_train)
y_pred_knn = knn.predict(x_test)
print("KNearest Neighbor Accuracy: ", knn.score(x_test, y_test))
```

KNearest Neighbor has the better accuracy. This was using k = 9 which, for me, returned the best accuracy score (3 = .941, 5 = .959, 7 = .964, 9 = .97).
