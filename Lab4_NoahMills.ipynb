```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
# Set the plot style to ggplot for a cleaner red/gray style.
plt.style.use('ggplot')

# Load the breast cancer dataset.
print("PART ONE (load data)")
data = load_breast_cancer()

# Create the pandas df from the above dataset.
# data.data is the numerical measurements of the dataset.
# data.feature_names is the columns names here (mean radius, mean texture, etc.)
print("\nPART TWO (created DataFrame)")
bcDataFrame = pd.DataFrame(data.data, columns=data.feature_names)

# Display the dataset into a pandas DataFrame.
# Display the DataFrames head & tail.
print("\nPART THREE")
print("\n", bcDataFrame.head())
print("\n", bcDataFrame.tail())

# Use the DataFrame method to calculate the descriptive statistics for
# the numerical data columns.
print("\nPART FOUR")
print("\n", bcDataFrame.describe())

# Use MatPlotLib to view histograms of each numerical data column.
print("\nPART FIVE")
# bcDataFrame automatically creates histograms for every numerical column.
# figsize=(12,8) alters the size of the histogram.
# bins=20 sets the number of bars (red blocks) the histograms have.
bcDataFrame.hist(figsize=(12,8), bins=20)
# Tight layout to avoid overlapping histograms.
plt.tight_layout()
```
